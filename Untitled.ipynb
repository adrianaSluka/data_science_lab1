### Вінтажний аналіз 
*Вінтажний аналіз*-частина поведінкового аналізу, суть якого полягає в аналізі поведінкових закономірностей однорідних груп клієнтів протягом їх життєвого циклу. <br>
Його необхідно провести, для того аби згенерувати мітки класів для кожного клієнта 

```python
import numpy as np
grouped = credit.groupby('ID')

### переведемо дані з таблиці credit у широкий формат, де кожен рядок відповідає унікальному ID

pivot_tb = credit.pivot(index = 'ID', columns = 'MONTHS_BALANCE', values = 'STATUS')
pivot_tb['open_month'] = grouped['MONTHS_BALANCE'].min() 

# open_month - мінімальний місяць у клієнта, місяць, коли було видано кредит

pivot_tb['end_month'] = grouped['MONTHS_BALANCE'].max()

# end_month - останній місяць активності клієнта

pivot_tb['ID'] = pivot_tb.index
pivot_tb = pivot_tb[['ID', 'open_month', 'end_month']]
pivot_tb['window'] = pivot_tb['end_month'] - pivot_tb['open_month'] # обрахуємо період спостереження
```

### Вибір вікна спостереження
Для оптимального аналізу даних і класифікації клієнтів, потрібно знайти найінформативніший період спостереження та відкинути записи, які не підходять для нього. Якщо вікно спостереження закоротке, поведінка клієнтів не буде повністю показана, що може бути причиною непотрібних шумів у даних

```python
lst = []
for i in range(0,61):
    ratio = len(pivot_tb[pivot_tb['window'] < i]) / len(set(pivot_tb['ID']))
    lst.append(ratio)
    
pd.Series(lst).plot(legend = False, grid = True, title = ' ')
plt.xlabel('Вікно спостереження')
plt.ylabel('Відносна кількість клієтів')
plt.show()
```

За графіком вище видно, що вікно спостереження у 60 місяців вміщає у себе 100 відсотків клієнтів.

```python
pivot_tb.reset_index(drop = True, inplace = True)
credit = pd.merge(credit, pivot_tb, on = 'ID', how = 'left') 

# додамо до таблиці credit обраховану вище інформацію

credit0 = credit.copy()
credit
```

```python
def calculate_observe(credit, command):
    '''calculate observe window
    '''
    id_sum = len(set(pivot_tb['ID']))
    credit['status'] = 0
    exec(command)
    credit['month_on_book'] = credit['MONTHS_BALANCE'] - credit['open_month']
    minagg = credit[credit['status'] == 1].groupby('ID')['month_on_book'].min()
    minagg = pd.DataFrame(minagg)
    minagg['ID'] = minagg.index
    obslst = pd.DataFrame({'month_on_book':range(0,61), 'rate': None})
    lst = []
    for i in range(0,61):
        due = list(minagg[minagg['month_on_book']  == i]['ID'])
        lst.extend(due)
        obslst.loc[obslst['month_on_book'] == i, 'rate'] = len(set(lst)) / id_sum 
    return obslst['rate']

command = "credit.loc[(credit['STATUS'] == '0') | (credit['STATUS'] == '1') | (credit['STATUS'] == '2') | (credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"   
morethan1 = calculate_observe(credit, command)
command = "credit.loc[(credit['STATUS'] == '1') | (credit['STATUS'] == '2') | (credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"   
morethan30 = calculate_observe(credit, command)
command = "credit.loc[(credit['STATUS'] == '2') | (credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"
morethan60 = calculate_observe(credit, command)
command = "credit.loc[(credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"
morethan90 = calculate_observe(credit, command)
command = "credit.loc[(credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"
morethan120 = calculate_observe(credit, command)
command = "credit.loc[(credit['STATUS'] == '5'), 'status'] = 1"
morethan150 = calculate_observe(credit, command)
```

```python
obslst = pd.DataFrame({'протерміновано більше ніж на 30 днів': morethan30,
                       'протерміновано більше ніж на 60 днів': morethan60,
                       'протерміновано більше ніж на 90 днів': morethan90,
                       'протерміновано більше ніж на 120 днів': morethan120,
                       'протерміновано більше ніж на 150 днів': morethan150
                        })
obslst.plot(grid = True, title = 'Кумулятивний відсоток "поганих" клієнтів')
plt.xlabel('Місяці')
plt.ylabel('Відсоток')
plt.show()
```

Даний графік можна трактувати наступним чином: наприклад, для категорії людей, які протермінували виплату кредиту більше ніж на 150 днів, потрібно мінімум 5 місяців, перш ніж перший "поганий" клієнт з'явиться; для більшості ситуацій, 20-місячне вікно спостереження покриває більшість ненадійних клієнтів. Тому пропоную обрати 20 місяців у якості нашого терміну спостереження, а всіх, у кого параметр window менше, необхідно виключити з аналізу.

## Вибір протермінування
Логічно, що протермінування більше 150 днів набагато гірше ніж в 30. Але необхідно обрати оптимальний проміжок часу для найбільше точного аналізу

```python
def calculate_rate(pivot_tb, command): 
    '''calculate bad customer rate
    '''
    credit0['status'] = None
    exec(command) # excuate input code
    sumagg = credit0.groupby('ID')['status'].agg(sum)
    pivot_tb = pd.merge(pivot_tb, sumagg, on = 'ID', how = 'left')
    pivot_tb.loc[pivot_tb['status'] > 1, 'status'] = 1
    rate = pivot_tb['status'].sum() / len(pivot_tb)
    return round(rate, 5)

command = "credit0.loc[(credit0['STATUS'] == '0') | (credit0['STATUS'] == '1') | (credit0['STATUS'] == '2') | (credit0['STATUS'] == '3' )| (credit0['STATUS'] == '4' )| (credit0['STATUS'] == '5'), 'status'] = 1"   
morethan1 = calculate_rate(pivot_tb, command)
command = "credit0.loc[(credit0['STATUS'] == '1') | (credit0['STATUS'] == '2') | (credit0['STATUS'] == '3' )| (credit0['STATUS'] == '4' )| (credit0['STATUS'] == '5'), 'status'] = 1"   
morethan30 = calculate_rate(pivot_tb, command)
command = "credit0.loc[(credit0['STATUS'] == '2') | (credit0['STATUS'] == '3' )| (credit0['STATUS'] == '4' )| (credit0['STATUS'] == '5'), 'status'] = 1"
morethan60 = calculate_rate(pivot_tb, command)
command = "credit0.loc[(credit0['STATUS'] == '3' )| (credit0['STATUS'] == '4' )| (credit0['STATUS'] == '5'), 'status'] = 1"
morethan90 = calculate_rate(pivot_tb, command)
command = "credit0.loc[(credit0['STATUS'] == '4' )| (credit0['STATUS'] == '5'), 'status'] = 1"
morethan120 = calculate_rate(pivot_tb, command)
command = "credit0.loc[(credit0['STATUS'] == '5'), 'status'] = 1"
morethan150 = calculate_rate(pivot_tb, command)

summary_dt = pd.DataFrame({'ситуація':['протерміновано більше ніж на 1 день',
                               'протерміновано більше ніж на 30 днів',
                               'протерміновано більше ніж на 60 днів',
                               'протерміновано більше ніж на 90 днів',
                               'протерміновано більше ніж на 120 днів',
                               'протерміновано більше ніж на 150 днів'],
                      'bad customer ratio':[morethan1,
                               morethan30,
                               morethan60,
                               morethan90, 
                               morethan120,
                               morethan150, 
                      ]})
summary_dt
```

Як видно по таблиці вище, 87% клієнтів мають протермінування у більше ніж 1 день, що є занадто частим для того, щоб бути стандартом. У свою чергу, протермінування більше ніж 150 днів мають всього 0.4%, що відкине забагато "поганих" клієнтів. Найкраще обрати протермінування у більше ніж 60 днів, що захопить 1.5% клієнтів, основна частина яких буде ненадійними.

```python
#видаляємо юзерів з терміном спостереження менше 20
credit = credit[credit['window'] > 20] 
```

```python
# обираємо записи, у яких протермінування більше 60 днів
credit['status'] = np.where((credit['STATUS'] == '2') | (credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 1, 0) 
credit['status'] = credit['status'].astype(np.int8) # 1: протерміновано 0: не протерміновано
```
